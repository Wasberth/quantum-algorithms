Una vez enviado un programa a la computadora cuántica de IBM, podemos obtener los resultados que tuvo con el ID que nos da la misma página de IBM.

Esto significa que podemos dejar corriendo (o esperando en la Queue), y no se interrumpe incluso si hacemos un Keyboard Interrupt.

Es decir, hay que interrumpirlo manualmente desde la página de IBM si queremos que no se ejecute

```python
from qiskit_ibm_runtime import QiskitRuntimeService
from qiskit.visualization import plot_histogram
```

```python
service = QiskitRuntimeService()

job = service.job('cvazjxka1h6g0089r0m0')
result = job.result()

print(result)
```

```python
counts = result.get_counts()
print(counts)
```

```python
plot_histogram(counts)
```
